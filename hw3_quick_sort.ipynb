```python
#這邊是一般的quick sort，沒有畫出排序狀況
import matplotlib.pyplot as plt

input_list = [33, 67, 8, 13, 54, 119, 3, 84, 25, 41]
print(input_list)

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0] #選擇比較基準值，設定為第一個元素
        left = [] #設定pivot左邊的列表，也就是比pivot小的數字們
        right = [] #設定pivot右邊的列表，也就是比pivot大的數字們
        for i in range(1, len(arr)):
            if arr[i] < pivot:
                left.append(arr[i])
            else:
                right.append(arr[i])
            final = quick_sort(left) + [pivot] + quick_sort(right)
        return final



arr = [33, 67, 8, 13, 54, 119, 3, 84, 25, 41]
quick_sort(input_list)

print(quick_sort(input_list))
```

```python
#這個是有畫出每一階段的排序狀況
import matplotlib.pyplot as plt

# 建立quicksort function
def quicksort(arr, left, right): #left為最左邊的index，right為最右邊的index
    if left < right:
        pivot = partition(arr, left, right)
        quicksort(arr, left, pivot - 1) #對左邊的子序列進行排序（遞迴）
        quicksort(arr, pivot + 1, right) #對右邊的子序列進行排序（遞迴)
    return arr


def partition(arr, left, right): #分割序列
    pivot = arr[right] #選擇基準值（pivot），設定為該序列最右邊的元素
    i = left - 1
    for j in range(left, right):
        if arr[j] <= pivot: #如果元素小於等於基準值，則選定該元素並進行交換
            i += 1
            arr[i], arr[j] = arr[j], arr[i]  #交換元素
        visualize_sort(arr, i, j, right, pivot) #畫出排序狀況
    arr[i + 1], arr[right] = arr[right], arr[i + 1]
    
    visualize_sort(arr, i + 1, right, right, pivot)
    return i + 1

'''
畫出排序狀況的function，這段程式碼部分參考：https://www.geeksforgeeks.org/visualization-of-quick-sort-using-matplotlib/
另一部分參考：https://youtu.be/iWQqm-vpbaw
'''

def visualize_sort(arr, i, j, pivot_idx, pivot_val): 

    fig, ax = plt.subplots()
    color = ['#1f77b4'] * len(arr) #設定長條為藍色
    color[i] = '#ff7f0e' #設定左邊的元素為橘色
    color[pivot_idx] = '#2ca02c' #設定基準值（pivot）為綠色
    ax.bar(range(len(arr)), arr, color=color)
    ax.set_xlim(0, len(arr))
    ax.set_ylim(0, max(arr))
    ax.set_title(f"QuickSort Algorithm (pivot={pivot_val})")
    for index, value in enumerate(arr):
        ax.text(index, value, str(value), ha = "center")
    plt.show(block=False)
    plt.pause(0.1)
    ax.clear()

def main():
    arr = [33, 67, 8, 13, 54, 119, 3, 84, 25, 41]
    visualize_sort(arr, -1, -1, -1, -1)
    quicksort(arr, 0, len(arr)-1)
    visualize_sort(arr, -1, -1, -1, -1)
    

if __name__ == "__main__":
    main()
```
